## B. Crawling Data Berita Website CNN

### 1. Install modul Beautifulsoup dan requests

### 2. Import library

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
````

### 3. Proses crawling menggunakan bs4 dan request

````python
#Inisialisasi Link URL
url = 'https://www.cnnindonesia.com/olahraga/sepakbola'

headers={
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
    }

#inisialisasi list untuk memasukkan hasil crawling kedalam list
listJudul = []
listIsi = []

#proses perulangan untuk crawling data
for page in range(1,4):
  req = requests.get(url+str(page), headers=headers)
  soup = BeautifulSoup(req.text, 'html.parser')
  items = soup.findAll('article')
  for it in items:
    try: link = it.find('a')['href']
    except : link=''
    try: title = it.find('h2', 'title').text 
    except : title=''
    
    if title != '':
      listJudul.append(title)

    if link != '':
      req2 = requests.get(str(link), headers=headers)
      soup2 = BeautifulSoup(req2.text, 'html.parser')
      items2 = soup2.findAll("div", "detail_text")
      for it2 in items2:
        try : berita = it2.find('p').text
        except : berita=''

        if berita != '':
          listIsi.append(berita)


df_Judul = pd.DataFrame(listJudul, columns=["Judul Berita"])
df_Isi = pd.DataFrame(listIsi, columns=["Isi Berita"])

data = pd.concat([df_Judul, df_Isi], axis=1)
data = data.loc[0:50]
data
````

## C. Crawling Data Jurnal di pta-trunojoyo.ac.id

*   Target data are ```["NIM", "Judul", "Abstrak", "Program Studi"]```

### 2. Import library

````python

import requests as req
from bs4 import BeautifulSoup as bs
from datetime import datetime
import snscrape.modules.twitter as sntwitter
import pandas as pd
import csv

userAgent = {'user-agent':'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Mobile Safari/537.36'}
````

### 3. Proses crawling data

````python
a = 1
for page in range(130, 170):
  url = f'https://pta.trunojoyo.ac.id/c_search/byprod/10/{page}'
  ge = req.get(url,userAgent).text
  sop = bs(ge, 'lxml')
  pd = sop.find('div',{'id':'begin'}).find('h2').text
  prodi = ''.join(pd).replace('Journal Jurusan','')
  li = sop.find('ul', class_='items list_style')
  lin = li.find_all('li')
  for x in lin:
    link = x.find('a', class_ ='gray button')['href']
    headline = x.find('div').find('a').text
    headline_ = ''.join(headline).replace('\n','')
    ge_ = req.get(link,userAgent).text
    sop_ = bs(ge_,'lxml')
    nim = x.find('div',{'style':'margin: 25px 0px 15px 0px;'}).find('a', class_='gray button')['href']
    nim_ = ''.join(nim).replace('https://pta.trunojoyo.ac.id/welcome/detail/','')
    content = sop_.find_all('ul', class_='items list_style')
    for x in content:
      nm = x.find('div',{'style':'padding:2px 2px 2px 2px;'}).find('span').text
      nama = ''.join(nm).replace('Penulis : ','')
      x = x.find_all('p')
      y = [y.text for y in x]
      content_ = ''.join(y).replace('\n','')
      print(f'done[{a}] >> {nim_[0:]}')
      a += 1
      with open('pta-trunojoyo.csv', 'a') as file:
        wr = csv.writer(file, delimiter=',')
        wr.writerow([nim_,nama,headline,content_,prodi])
````
